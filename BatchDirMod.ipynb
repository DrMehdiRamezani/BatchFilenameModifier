```python
import os

def batch_rename_photos(directory_path, new_prefix):
    # List all files in the directory
    files = os.listdir(directory_path)

    # Filter only files with specific extensions (e.g., '.jpg', '.png')
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

    # Iterate through the image files and rename them
    for index, old_name in enumerate(image_files, start=1):
        # Generate the new file name
        file_extension = os.path.splitext(old_name)[1]
        new_name = f"{new_prefix}_{index}{file_extension}"

        # Build the full paths for old and new names
        old_path = os.path.join(directory_path, old_name)
        new_path = os.path.join(directory_path, new_name)

        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {old_name} -> {new_name}")

# Replace 'your_directory_path' with the actual path to your photo directory
# Replace 'new_photo_name_prefix' with the desired prefix for the new names
batch_rename_photos('/home/mehdi/data_dir/temperature/', 'temperature_icon_')
```
